defining function to read url

```python
import requests
import json
import urllib

def get_symbol(name):
    url = "http://d.yimg.com/autoc.finance.yahoo.com/autoc?query={}&region=1&lang=en".format(name)

    result = requests.get(url).json()
    
    for x in result['ResultSet']['Result']:
             return x['symbol']
        
ticker=[]
company = df.ENTITY
    
for  xyz in company:
    ticker.append(get_symbol(xyz))
    
    
print(ticker
     )
```

```python
import datetime as dt
start = dt.datetime(2015,1,1)
end = dt.datetime(2017,6,6)
data_source = 'yahoo'
df1 = pdr.DataReader(ticker, data_source, start, end)
```

Re-indexing the stocks with the required time period

```python

close = df1.ix['Close']
days = pd.date_range(start=start, end=end, freq='B')
close = close.reindex(days)
close
```
